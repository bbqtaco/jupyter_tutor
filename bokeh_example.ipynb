```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d.axes3d import Axes3D
%matplotlib notebook
```

```python
df = pd.read_csv('datasets/iris.csv',index_col=4,header=None,names=['SepalLength', 'SepalWidth', 'PetalLength','PetalWidth'],sep=',')
df.head()
x=df['SepalLength']
y=df['SepalWidth']
z=df['PetalLength']
```

```python
df1 = df[df.index == 'Iris-setosa']
x1=df1['SepalLength']
y1=df1['SepalWidth']
z1=df1['PetalLength']
df2 = df[df.index == 'Iris-virginica']
x2=df2['SepalLength']
y2=df2['SepalWidth']
z2=df2['PetalLength']
df3 = df[df.index == 'Iris-versicolor']
x3=df3['SepalLength']
y3=df3['SepalWidth']
z3=df3['PetalLength']
```
